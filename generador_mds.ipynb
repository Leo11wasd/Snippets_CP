```python
from pathlib import Path

ROOT = Path(".")  # change to your snippets directory if needed
template = Path("template.md").read_text(encoding="utf-8")

for cpp_file in ROOT.rglob("*.cpp"):
    md_file = cpp_file.with_suffix(".md")

    # Skip if markdown file already exists
    if md_file.exists():
        print(f"Skipping {md_file}")
        continue

    code = cpp_file.read_text(encoding="utf-8")

    content = template.format(
        name=cpp_file.stem,
        code=code.rstrip()
    )

    md_file.write_text(content, encoding="utf-8")
```
